```python
import numpy as np
import matplotlib.pyplot as plt

import librosa
import librosa.display

import IPython
```

```python
data = np.load("/tf/data/4stem_arr.npz", allow_pickle=True)
```

```python
index = 100
example_song = data["arr_0"][index]
example_label = data["arr_1"][index]

data["arr_0"][index].keys()
print(f"Label {example_label}")
```

```python
fig_1, ax_1 = plt.subplots()
img_1 = librosa.display.specshow(example_song["original"], x_axis='time', y_axis='linear', ax=ax_1)
ax_1.set(title="Original")
fig_1.colorbar(img_1, ax=ax_1, format="%+2.f dB")

fig_2, ax_2 = plt.subplots()
img_2 = librosa.display.specshow(example_song["vocals"], x_axis='time', y_axis='linear', ax=ax_2)
ax_2.set(title="Vocals")
fig_2.colorbar(img_2, ax=ax_2, format="%+2.f dB")

fig_3, ax_3 = plt.subplots()
img_3 = librosa.display.specshow(example_song["drums"], x_axis='time', y_axis='linear', ax=ax_3)
ax_3.set(title="Drums")
fig_3.colorbar(img_3, ax=ax_3, format="%+2.f dB")

fig_4, ax_4 = plt.subplots()
img_4 = librosa.display.specshow(example_song["bass"], x_axis='time', y_axis='linear', ax=ax_4)
ax_4.set(title="Bass")
fig_4.colorbar(img_4, ax=ax_4, format="%+2.f dB")

fig_5, ax_5 = plt.subplots()
img_5 = librosa.display.specshow(example_song["other"], x_axis='time', y_axis='linear', ax=ax_5)
ax_5.set(title="Other")
fig_5.colorbar(img_5, ax=ax_5, format="%+2.f dB")
```
