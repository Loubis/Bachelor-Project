```python
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('/home/steffen/Datasets/prepro/arr_2.npy', allow_pickle=True)
data[0][1][1]
```

```python
for spect in data[0][1][1]:
    fig, ax = plt.subplots()
    img = librosa.display.specshow(spect[0], x_axis='time', y_axis='log', ax=ax)
    fig.colorbar(img, ax=ax, format="%+2.f dB")
    
```
